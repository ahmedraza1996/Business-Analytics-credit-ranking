# Single Sample Mean ( Sigma Known) 

### H0 :  the mean age of Population is 36
### H1 : the mean age of Population is not 36 


```python
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(df['Age'], x2=None, value=36 ,alternative='two-sided')
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
