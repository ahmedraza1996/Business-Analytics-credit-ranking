```python
def style_specific_cell(x):

    color_thresh = 'background-color: lightpink'
    
    df_color = pd.DataFrame('', index=x.index, columns=x.columns)
    rows_number=len(x.index)
    column_number=len(x.columns)
    for r in range(0,rows_number): 
        for c in range(0,column_number):
            try:
                val=float(x.iloc[r, c])
                if x.iloc[r, 0]=="Percentage":
                    if val<10:
                        df_color.iloc[r, c]=color_thresh
            except:
                pass
            
    return df_color

def style_stats_specific_cell(x):

    color_thresh = 'background-color: lightpink'
    
    df_color = pd.DataFrame('', index=x.index, columns=x.columns)
    rows_number=len(x.index)
    for r in range(0,rows_number):
        try:
            val=(x.iloc[r, 1])
            if val>0.05:
                df_color.iloc[r, 1]=color_thresh
        except:
            pass
    return df_color
```

# Hypothesis 2: 

### H0: Defaultors have a mean value of Credit Amount greater than 7500
### H1:  Defaultors have a mean value of Credit Amount not greater than 7500

```python
df_credit=pd.DataFrame(df[df["CreditAmount"]>=7500])
```

```python
df_credit.reset_index()
```

```python
df_Bad=pd.DataFrame(df[df["Class"]==2])
```

```python
df_Bad.reset_index()
```

```python
df["CreditAmount"]=(df["CreditAmount"]-df["CreditAmount"].mean())/df["CreditAmount"].std()
```

```python
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(df_Bad['CreditAmount'], x2=None, value=7500)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

#### So we reject the null hypothesis "Defaultors have a mean value of Credit Amount greater than 7500"

# Hypothesis 3: 

### H0 : there is mean difference between credit amount of Good and Bad
### H1 : there is no mean difference between credit amount of Good and Bad

```python
bad=pd.DataFrame(df[df["Class"]==2])
good=pd.DataFrame(df[df["Class"]==1])
```

```python
ztest ,pval = stests.ztest(bad['CreditAmount'], x2=good['CreditAmount'], value=0,alternative='two-sided')
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

#### So, there is a difference between mean credit amount of Good and Bad

# ONE HOT ENCODING OF CATEGORICAL VARIABLES

```python
col_cat_names=["checkingAccount","CreditHistory","Purpose",\
"SavingAccounts","PersonalStatus","PresentEmployment","Property",
"OtherInstallmentPlans","Housing","ForeignWorker"]
for attr in col_cat_names:
    df = df.merge(pd.get_dummies(df[attr], prefix=attr), left_index=True, right_index=True)
    df.drop(attr,axis=1,inplace=True)
```

# APPLYING MODEL

## METHOD 1: USING XGBOOST

```python
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix,roc_curve
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
y = df["Class"].values #target variables
X = df.drop('Class', 1).values #independent variables
pca = PCA(n_components=16)
X = pca.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import KFold # import KFold


kf = KFold(n_splits=5) # Define the split - into 2 folds 
kf.get_n_splits(X) # returns the number of splitting iterations in the cross-validator
print(kf) 
```

```python
xgb_model = XGBClassifier()
```

```python
count=1
for train_index, test_index in kf.split(X):
    
    print("\n FOLD: "+str(count)+ "\n")
    # print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    xgb_model.fit(X_train, y_train)
    print('Accuracy = ',accuracy_score(y_test, xgb_model.predict(X_test)))
    print(classification_report(y_test, xgb_model.predict(X_test)))
    print("Confusion Matrix: \n",confusion_matrix(y_test,  
                                                  xgb_model.predict(X_test), labels=[1, 2]))
    count+=1
```
